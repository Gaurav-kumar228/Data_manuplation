```
df1filtered_years = df1[
    (df1['time'].dt.year >= 1990) &
    (df1['time'].dt.year <= 2022)
].copy()
print(df1filtered_years)
```

```
df_summer_max_temp = df1filtered_years[
    df1filtered_years['time'].dt.month.isin([3, 4, 5])
]
print(df_summer_max_temp)
```

```
df_monsoon_max_temp = df1filtered_years[
    df1filtered_years['time'].dt.month.isin([6, 7, 8, 9])
]
print(df_monsoon_max_temp)
print(len(df_monsoon_max_temp))
```

```
df_autumn_max_temp = df1filtered_years[
    df1filtered_years['time'].dt.month.isin([10, 11])
]
print(df_autumn_max_temp)
print(len(df_autumn_max_temp))
```

```
df_winter_max_temp = df1filtered_years[
    df1filtered_years['time'].dt.month.isin([12, 1, 2])
]
print(df_winter_max_temp)
```

```

df1['year'] = df1['time'].dt.year
df1['month'] = df1['time'].dt.month

# Assign seasons
def get_season(month):
    if month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    elif month in [10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

df1['season'] = df1['month'].apply(get_season)

# Calculate mean temperature per year per season
seasonal_mean = (
    df1
    .groupby(['year', 'season'])['tmax']
    .mean()
    .rename('Mean Max Temperature') # Explicitly rename the aggregated column
    .reset_index()
)

print(seasonal_mean.head())

# Pivot for plotting
pivot_df = seasonal_mean.pivot(
    index='year',
    columns='season',
    values='Mean Max Temperature' # Use the explicitly renamed column
)

# Plot
plt.figure(figsize=(12,6))

for season in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[season], label=season)

plt.xlabel('Year')
plt.ylabel('Mean Maximum Temperature')
plt.title('Season-wise Mean Maximum Temperature (19902022)')
plt.legend()
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(12,6))
for season in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[season], label=f"{season} Mean")

years = pivot_df.index.values

for season in pivot_df.columns:
    temps = pivot_df[season].values


    mask = ~np.isnan(temps)
    x = years[mask]
    y = temps[mask]

    m, c = np.polyfit(x, y, 1)

    y_pred = m * x + c

    plt.plot(x, y_pred, linestyle='--', label=f"{season} Trend")

plt.xlabel('Year')
plt.ylabel('Mean Maximum Temperature')
plt.title('Season-wise Mean Temperature with Regression Trends (19902022)')
plt.legend()
plt.grid(True)

plt.show()
```

```
# Calculate variance per year per season
seasonal_variance = (
    df_max_temp
    .groupby(['year', 'season'])['tmax']
    .var()
    .reset_index()
)

print(seasonal_variance.head())
```

```
variance_pivot = seasonal_variance.pivot(
    index='year',
    columns='season',
    values='tmax'
)
print(variance_pivot)
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for season in variance_pivot.columns:
    plt.plot(variance_pivot.index, variance_pivot[season], label=season)

plt.xlabel('Year')
plt.ylabel('Temperature Variance')
plt.title('Season-wise Temperature Variance (19902022)')
plt.legend()
plt.grid(True)
plt.show()
```
